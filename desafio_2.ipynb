```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df_aportes_elecciones = pd.read_excel("input/aportes_elecciones_2017.xlsx")
df_aportes_elecciones["ref"] = 1
df_aportes_elecciones
```

# matplotlib

```python
mpl.style.use('bmh')
```

```python
# Cantidad de aportes por candidato presidencial
fig, ax = plt.subplots(figsize=(20,5))

df_monto_candidato = df_aportes_elecciones[["MONTO"]].groupby(df_aportes_elecciones[df_aportes_elecciones["ELECCION"] == "PRESIDENTE"]["NOMBRE CANDIDATO/PARTIDO POLITICO"]).sum()
x = df_monto_candidato.index
y = df_monto_candidato.MONTO.values

ax.bar(x, y, label='Monto total')
ax.set_title('Monto por candidato o partido político')
ax.set_xlabel("Candidato o partido político")
ax.set_ylabel("Monto")

plt.xticks(rotation=45)
ax.legend()
plt.show()
```

```python
# En que fecha se realizan mas aportes en candidaturas presidenciales
fig, ax = plt.subplots(figsize=(20,5))

df_fecha_monto = df_aportes_elecciones[["MONTO"]].groupby(df_aportes_elecciones[df_aportes_elecciones["ELECCION"] == "PRESIDENTE"]["FECHA DE\nTRANSFERENCIA"]).sum()

fecha = df_fecha_monto.index
monto = df_fecha_monto.MONTO.values

ax.scatter(fecha, monto, label='Monto total')
ax.set_title('Mayor contidad de aportes por fecha')
ax.set_xlabel('Fecha')
ax.set_ylabel('Monto')

ax.legend()
plt.show()
```

```python
# Cantidad de aportes publicos y privados para candidatura presidencial
df_tipo_aporte = df_aportes_elecciones[["ref"]].groupby(df_aportes_elecciones[df_aportes_elecciones["ELECCION"] == "PRESIDENTE"]["TIPO DE\n APORTE"]).sum()

tipo_aporte = df_tipo_aporte.index
cant_aportes = df_tipo_aporte.ref.values

fig,ax = plt.subplots(figsize =(10, 7))
ax.pie(cant_aportes, labels = tipo_aporte)

plt.show()
```

# seaborn

```python
# Cantidad de aportes por candidato presidencial
fig, ax = plt.subplots(figsize=(20,5))

df_monto_candidato = df_aportes_elecciones[["MONTO"]].groupby(df_aportes_elecciones[df_aportes_elecciones["ELECCION"] == "PRESIDENTE"]["NOMBRE CANDIDATO/PARTIDO POLITICO"]).sum()
x = df_monto_candidato.index
y = df_monto_candidato.MONTO.values

sns.barplot(data=df_monto_candidato, x=x, y=y)

ax.set_title('Monto por candidato o partido político')
ax.set_xlabel("Candidato o partido político")
ax.set_ylabel("Monto")

plt.xticks(rotation=45)
plt.show()
```

```python
# En que fecha se realizan mas aportes en candidaturas presidenciales
fig, ax = plt.subplots(figsize=(20,5))

df_fecha_monto = df_aportes_elecciones[["MONTO"]].groupby(df_aportes_elecciones[df_aportes_elecciones["ELECCION"] == "PRESIDENTE"]["FECHA DE\nTRANSFERENCIA"]).sum()

fecha = df_fecha_monto.index
monto = df_fecha_monto.MONTO.values

sns.scatterplot(data=df_fecha_monto, x=fecha, y=monto)
ax.set_title('Mayor contidad de aportes por fecha')
ax.set_xlabel('Fecha')
ax.set_ylabel('Monto')

plt.show()
```

```python
# Cantidad de aportes publicos y privados para candidatura presidencial
df_tipo_aporte = df_aportes_elecciones[["ref"]].groupby(df_aportes_elecciones[df_aportes_elecciones["ELECCION"] == "PRESIDENTE"]["TIPO DE\n APORTE"]).sum()

tipo_aporte = df_tipo_aporte.index
cant_aportes = df_tipo_aporte.ref.values

fig,ax = plt.subplots(figsize =(10, 7))

colors = sns.color_palette('pastel')[0:5]

ax.pie(cant_aportes, labels = tipo_aporte, colors = colors, autopct='%.0f%%')
plt.show()
```
